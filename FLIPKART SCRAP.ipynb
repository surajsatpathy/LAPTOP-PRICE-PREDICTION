```python
productName=[]
processor=[]
ram=[]
rating=[]
ssd=[]
os=[]
screensize=[]
price=[]
mrp=[]
for k in range(1,41):
    url='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}'.format(k)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML ,like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    container=soup.find_all('div',class_='_2kHMtA')
    for i in container:
        name=i.find('div',class_='_4rR01T')
        productName.append(name.text.split('-')[0].strip())
        p=i.find_all('li',class_='rgWa7D')
        processor.append(p[0].text)
        
        if ('GB') in i.find_all('li',class_='rgWa7D')[1].text:
            ram.append(i.find_all('li',class_='rgWa7D')[1].text)
        else:
            ram.append('nan')
        rat=i.find('div',class_='_3LWZlK')
        try:
            rating.append(rat.text)
        except:
            rating.append('0')
        try:
            if ('Operating') in i.find_all('li',class_='rgWa7D')[2].text:
                os.append(i.find_all('li',class_='rgWa7D')[2].text)
            else:
                os.append('nan')
        except:
            os.append('No os')
        try:
            if ('SSD') in  i.find_all('li',class_='rgWa7D')[3].text:
                ssd.append(i.find_all('li',class_='rgWa7D')[3].text)
            else:
                ssd.append('nan')
        except:
            ssd.append('nan')
        if ('cm') in i.find_all('li',class_='rgWa7D')[4].text:
            screensize.append(i.find_all('li',class_='rgWa7D')[4].text)
        else:
            screensize.append('nan')
        pr=i.find_all('div',class_='_30jeq3 _1_WHN1')
        price.append(pr[0].text)
        mr=i.find_all('div',class_='_3I9_wc _27UcVY')
        try:
            mrp.append(mr[0].text)
        except:
            mrp.append('nan')
```

```python
db={}
```

```python
db['productName']=productName
db['Processor']=processor
db['Ram']=ram
db['Rating']=rating
db['SSD']=ssd
db['OS']=os
db['screensize']=screensize
db['Price']=price
db['Mrp']=mrp
```

```python
df = pd.DataFrame(db)
```

```python
df=pd.read_csv('final.csv')
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X = df.drop(columns=['Price'])
y = np.log(df['Price'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first'),[0,2,5])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
